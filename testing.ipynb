```python
from pathlib import Path
import os
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
data_dir = Path("data/treemap/segs_w_species")
files = os.listdir(data_dir)

all_data = []
for i,file in enumerate(files):
    print(f"Processing file {i}")
    all_data.append(gpd.read_file(data_dir/file))

all_data = pd.concat(all_data, axis = 0)
```

```python
all_data.to_file(data_dir/"treemap.gpkg", driver = "GPKG")
```
